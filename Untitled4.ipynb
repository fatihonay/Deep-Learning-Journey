# Tensor Operations
Let's do some tensor operations to get much more familiarity with basics.

```python
import torch
```

```python
tensor_1D = torch.arange(16)
tensor_1D
```

We can change and manipulate the shape of tensors as we wish.

```python
tensor_2D = tensor_1D.reshape(2,8)
tensor_2D
```

Sometime we will have to deal with large dimensions and not to know exact number of rows and columns. We can still reshape dimensions easily. "-1" will provide this functionality.

```python
tensor_2D = tensor_1D.reshape(4,-1)
tensor_2D
```

We created and reshaped tensors so far. But there is more than that. Let's change conctent of the tensors.


```python
tensor_x = torch.arange(15)
tensor_y = tensor_x.reshape(3,-1) # Automatically distribute columns

print('Tensor X =',tensor_x)
print('Tensor Y =',tensor_y)

# Change content
tensor_x[-1] = 99 # Change last element

print('Tensor X =',tensor_x)

print(10*'#','Tensor Y Automaticaly updated',10*'#')
print('Tensor Y =',tensor_y)
```

So far so good! Everything is so easy, hah. Let's do some slicing

```python
tensor_z = torch.arange(12)
tensor_w = tensor_z.reshape(-1,2)

# Print 2D tensor
print(tensor_w)

# Choose column index 1
print(tensor_w[:,1] )

# Choose column index 1 and get output in column format using :
tensor_w[:,1:]
```
