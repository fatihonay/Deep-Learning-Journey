# Lists, Arrays, Tensors

We will study basics of tensors with PyTorch for deep learning. Tensors are similar to Numpy arrays which are pretty similar to Python lists.

Unlike Numpy Arrays, Tensors are built to be used with GPU's for Deep Learning.

```python
import torch
import numpy as np
```

### Lists
The primary role of a list is to hold multiple items in a single variable.
The first item is always at index 0, the second at index 1, etc. Python lists can store mixed data types (integers, strings, objects, and even other lists) together.

```python
odd_list = [1,3,5,7,9]
even_list = [0,2,4,6,8]

even_list
```

### Arrays

Numpyy arrays are specialized data structures designed for scientific computing and high-performance data analysis. The most critical role of a NumPy array is vectorization. This allows to perform mathematical operations on entire arrays at once without writing loops.

```python
array_rand = np.random.rand(5,4)
array_rand
```

Let's check data type

```python
print(array_rand.dtype)
```

### Tensors

In the context of Python and data science, Tensors are the next evolution after lists and NumPy arrays. They are the fundamental data structure used in Deep Learning frameworks. NumPy arrays live on the CPU but Tensors can be moved to the GPU. And this increase processing speed dramatically for training deep learning models.


```python
tensor_rand = torch.rand(5,4)
tensor_rand
```

As you can see, Tensors are Numpy Arrays are so similar. We can also convert Numpy arrays into Tensors.

```python
arra_to_tensor = torch.tensor(array_rand)
arra_to_tensor
```
